```python
import numpy as np
import matplotlib.pyplot as plt

# Define the objective function
def objective_function(x, y):
    return (x - 1)**2 + (y - 2.5)**2

# Create a grid for x and y
x = np.linspace(-1, 5, 400)
y = np.linspace(-1, 5, 400)
X, Y = np.meshgrid(x, y)
Z = objective_function(X, Y)

# Define the constraints
constraint1 = X - 2*Y + 2 >= 0
constraint2 = -X - 2*Y + 6 >= 0
constraint3 = -X + 2*Y + 2 >= 0
constraint4 = X >= 0
constraint5 = Y >= 0

# Combine all constraints
feasible_region = constraint1 & constraint2 & constraint3 & constraint4 & constraint5

# Plot the level sets of the objective function
plt.figure(figsize=(8, 8))
contour = plt.contour(X, Y, Z, levels=25, cmap='coolwarm')
plt.clabel(contour, inline=1, fontsize=8)

# Highlight the feasible region
plt.imshow(feasible_region, extent=(-1, 5, -1, 5), origin='lower', alpha=0.3, cmap='Greens')

# Draw constraint boundaries
plt.plot(x, 0.5*x - 1, 'k-', label='x - 2y + 2 = 0')
plt.plot(x, -0.5*x + 3, 'b-', label='-x - 2y + 6 = 0')
plt.plot(x, 0.5*x + 1, 'g-', label='-x + 2y + 2 = 0')
plt.axvline(0, color='purple', linestyle='--', label='x = 0')
plt.axhline(0, color='orange', linestyle='--', label='y = 0')

# Labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Level Sets of Objective Function and Feasible Region')
plt.legend()
plt.grid()
plt.show()
```

```python
# Reimport necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the objective function
def objective_function(x, y):
    return (x - 1)**2 + (y - 2.5)**2

# Create a grid for x and y
x = np.linspace(-1, 5, 400)
y = np.linspace(-1, 5, 400)
X, Y = np.meshgrid(x, y)
Z = objective_function(X, Y)

# Define the constraints
constraint1 = X - 2*Y + 2 >= 0
constraint2 = -X - 2*Y + 6 >= 0
constraint3 = -X + 2*Y + 2 >= 0
constraint4 = X >= 0
constraint5 = Y >= 0

# Combine all constraints
feasible_region = constraint1 & constraint2 & constraint3 & constraint4 & constraint5

# Coordinates for iteration points
iteration_points = [(2, 0), (2.2, 0.1), (1.4, 1.7)]
labels = [
    "Iter 1: (2, 0), W={3, 5}",
    "Iter 2: (2.2, 0.1), W={3}",
    "Iter 4: (1.4, 1.7), W="
]

# Plot level sets and feasible region
plt.figure(figsize=(8, 8))
contour = plt.contour(X, Y, Z, levels=20, cmap='coolwarm')
plt.clabel(contour, inline=1, fontsize=8)
plt.imshow(feasible_region, extent=(-1, 5, -1, 5), origin='lower', alpha=0.3, cmap='Greens')

# Plot iteration points and annotate
for i, (point, label) in enumerate(zip(iteration_points, labels)):
    plt.plot(point[0], point[1], 'ro')  # Mark iteration point
    plt.text(point[0] + 0.1, point[1], label, fontsize=8, color='red')

# Draw constraint boundaries
plt.plot(x, 0.5*x - 1, 'k-', label='x - 2y + 2 = 0')
plt.plot(x, -0.5*x + 3, 'b-', label='-x - 2y + 6 = 0')
plt.plot(x, 0.5*x + 1, 'g-', label='-x + 2y + 2 = 0')
plt.axvline(0, color='purple', linestyle='--', label='x = 0')
plt.axhline(0, color='orange', linestyle='--', label='y = 0')

# Highlight the final solution
plt.plot(1.4, 1.7, 'mo', markersize=8, label='Optimal Solution (1.4, 1.7)')

# Labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Level Sets, Feasible Region, and Iteration Points')
plt.legend()
plt.grid()
plt.show()
```
